```python
import pandas as pd

# DECP
decp = pd.read_csv(
    "data/raw/decp-2022-marches-valides.csv",
    sep=";",
    low_memory=False,
    on_bad_lines='skip'
)
```

```python
print("decp :", decp.shape)

decp.head()
```

```python
print("Colonnes DECP :", decp.columns.tolist())
```

```python
print("decp", decp.isna().sum().sort_values(ascending=False).head(10))
```

```python
# Vérifier les colonnes d'identifiants dans DECP
print("DECP - taux de remplissage")
print("acheteur.id :", decp["acheteur.id"].notna().mean())
print("titulaire_id_1 :", decp["titulaire_id_1"].notna().mean())
print("titulaire_id_2 :", decp["titulaire_id_2"].notna().mean())
print("titulaire_id_3 :", decp["titulaire_id_3"].notna().mean())
```

```python
decp["acheteur.id"] = decp["acheteur.id"].astype(str)
decp["titulaire_id_1"] = decp["titulaire_id_1"].astype(str)

decp = decp.rename(columns={
    "acheteur.id": "acheteur_id",
    "titulaire_id_1": "titulaire_id"
})

# Vérifier
print(decp[["acheteur_id", "titulaire_id"]].head())
```

```python
# Transformer les SIRET en SIREN dans DECP
decp["acheteur_id"] = decp["acheteur_id"].str[:9]
decp["titulaire_id"] = decp["titulaire_id"].str[:9]

# Vérifier
print(decp[["acheteur_id", "titulaire_id"]].head())
```

```python
# Sélection des colonnes utiles
decp_clean = decp[[
    "acheteur_id",
    "titulaire_id",
    "montant",
    "dureeMois",
    "dateNotification",
    "procedure",
    "codeCPV",
    "objet",
    "marcheInnovant",
    "considerationsSociales",
    "considerationsEnvironnementales"
]].copy()

# Convertir dateNotification en datetime
decp_clean["dateNotification"] = pd.to_datetime(decp_clean["dateNotification"], errors='coerce')

# Vérifier le nettoyage
decp_clean.info()
decp_clean.head()
```

```python
# ==========================
# KPIs financiers et qualitatifs
# ==========================

import pandas as pd

# 1️⃣ Montant total par acheteur
montant_par_acheteur = decp_clean.groupby("acheteur_id")["montant"].sum().sort_values(ascending=False)
print("🔹 Top 10 acheteurs par montant total :")
display(montant_par_acheteur.head(10))

# 2️⃣ Nombre de marchés par type de procédure
marches_par_procedure = decp_clean.groupby("procedure")["montant"].count().sort_values(ascending=False)
print("\n🔹 Nombre de marchés par type de procédure :")
display(marches_par_procedure)

# 3️⃣ Durée moyenne des marchés
duree_moyenne = decp_clean["dureeMois"].mean()
print("\n🔹 Durée moyenne des marchés (mois) :", round(duree_moyenne, 2))

# 4️⃣ Proportion de marchés innovants
taux_innovant = (decp_clean["marcheInnovant"].str.lower() == "oui").mean()
print("\n🔹 Proportion de marchés innovants :", round(taux_innovant*100, 2), "%")

# 5️⃣ Proportion de marchés avec critères sociaux / environnementaux
taux_social = (decp_clean["considerationsSociales"].str.lower() != "pas de considération sociale").mean()
taux_env = (decp_clean["considerationsEnvironnementales"].str.lower() != "pas de considération environnementale").mean()
print("🔹 Proportion de marchés avec critères sociaux :", round(taux_social*100, 2), "%")
print("🔹 Proportion de marchés avec critères environnementaux :", round(taux_env*100, 2), "%")

# ==========================
# Bonus : Top 10 acheteurs par montant et nombre de marchés
# ==========================
top_acheteurs = decp_clean.groupby("acheteur_id").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False)

print("\n🔹 Top 10 acheteurs combiné montant et nombre de marchés :")
display(top_acheteurs.head(10))
```

```python
# ==========================
# Mini-dashboard KPIs DECP
# ==========================
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 1️⃣ Top 10 acheteurs par montant
top10_montant = montant_par_acheteur.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top10_montant.values, y=top10_montant.index, palette="viridis")
plt.title("Top 10 acheteurs par montant total (€)")
plt.xlabel("Montant total")
plt.ylabel("Acheteur ID")
plt.show()

# 2️⃣ Nombre de marchés par type de procédure (Top 10)
top10_procedure = marches_par_procedure.head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top10_procedure.values, y=top10_procedure.index, palette="magma")
plt.title("Nombre de marchés par type de procédure (Top 10)")
plt.xlabel("Nombre de marchés")
plt.ylabel("Procédure")
plt.show()

# 3️⃣ Proportion de marchés innovants / sociaux / environnementaux
kpi_percentages = {
    "Innovants": taux_innovant*100,
    "Avec critères sociaux": taux_social*100,
    "Avec critères environnementaux": taux_env*100
}

plt.figure(figsize=(8,6))
sns.barplot(x=list(kpi_percentages.keys()), y=list(kpi_percentages.values()), palette="coolwarm")
plt.title("Proportion de marchés par type de KPI (%)")
plt.ylabel("Pourcentage")
plt.ylim(0,100)
plt.show()

# 4️⃣ Distribution de la durée des marchés
plt.figure(figsize=(10,6))
sns.histplot(decp_clean["dureeMois"], bins=50, kde=True, color="skyblue")
plt.title("Distribution de la durée des marchés (mois)")
plt.xlabel("Durée (mois)")
plt.ylabel("Nombre de marchés")
plt.show()
```

```python
# Agréger par acheteur
top_acheteurs_dash = decp_clean.groupby("acheteur_id").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False).head(10)

# Graphique combiné
plt.figure(figsize=(12,6))
sns.barplot(x=top_acheteurs_dash.index, y=top_acheteurs_dash["total_montant"], color="skyblue", label="Montant total (€)")
sns.lineplot(x=top_acheteurs_dash.index, y=top_acheteurs_dash["nb_marches"], color="red", marker="o", label="Nombre de marchés")
plt.title("Top 10 acheteurs : Montant total vs Nombre de marchés")
plt.xlabel("Acheteur ID")
plt.ylabel("Montant total")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# Top 10 acheteurs par montant total et nombre de marchés (version texte)
print("🔹 Top 10 acheteurs (texte) :")
display(top_acheteurs_dash)

```

```python
# Top 10 codes CPV par montant total
top_cpv = decp_clean.groupby("codeCPV").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False).head(10)

# Graphique combiné
plt.figure(figsize=(12,6))
sns.barplot(x=top_cpv.index, y=top_cpv["total_montant"], color="lightgreen", label="Montant total (€)")
sns.lineplot(x=top_cpv.index, y=top_cpv["nb_marches"], color="orange", marker="o", label="Nombre de marchés")
plt.title("Top 10 secteurs (code CPV) : Montant total vs Nombre de marchés")
plt.xlabel("Code CPV")
plt.ylabel("Montant total")
plt.xticks(rotation=45)
plt.legend()
plt.show()

print("\n🔹 Top 10 secteurs (code CPV) (texte) :")
display(top_cpv)
```

```python
# ==========================
# Analyse temporelle et KPIs qualitatifs
# ==========================

# Convertir dateNotification en datetime si ce n'est pas déjà fait
decp_clean["dateNotification"] = pd.to_datetime(decp_clean["dateNotification"], errors='coerce')

# Extraire l'année
decp_clean["annee"] = decp_clean["dateNotification"].dt.year

# 1️⃣ Montant total et nombre de marchés par année
montant_par_annee = decp_clean.groupby("annee")["montant"].sum()
nb_marches_par_annee = decp_clean.groupby("annee")["montant"].count()

print("🔹 Montant total par année :")
display(montant_par_annee)

print("\n🔹 Nombre de marchés par année :")
display(nb_marches_par_annee)

# Graphique combiné
plt.figure(figsize=(12,6))
sns.lineplot(x=montant_par_annee.index, y=montant_par_annee.values, marker="o", label="Montant total (€)")
sns.lineplot(x=nb_marches_par_annee.index, y=nb_marches_par_annee.values, marker="o", label="Nombre de marchés")
plt.title("Évolution du montant total et du nombre de marchés par année")
plt.xlabel("Année")
plt.ylabel("Valeur")
plt.legend()
plt.show()

# 2️⃣ KPIs qualitatifs par année
kpi_annee = decp_clean.groupby("annee").agg(
    taux_innovant=pd.NamedAgg(column="marcheInnovant", aggfunc=lambda x: (x.str.lower() == "oui").mean()),
    taux_social=pd.NamedAgg(column="considerationsSociales", aggfunc=lambda x: (x.str.lower() != "pas de considération sociale").mean()),
    taux_env=pd.NamedAgg(column="considerationsEnvironnementales", aggfunc=lambda x: (x.str.lower() != "pas de considération environnementale").mean())
)

print("\n🔹 KPIs qualitatifs par année :")
display(kpi_annee)

# Graphique des KPI qualitatifs
plt.figure(figsize=(12,6))
sns.lineplot(data=kpi_annee*100, markers=True)
plt.title("Évolution des KPI qualitatifs par année (%)")
plt.ylabel("Pourcentage (%)")
plt.xlabel("Année")
plt.legend(["Innovants", "Critères sociaux", "Critères environnementaux"])
plt.show()
```

```python
# ==========================
# Mini-dashboard final DECP
# ==========================

import matplotlib.pyplot as plt
import seaborn as sns

# Filtrer les années réalistes
decp_dash = decp_clean[(decp_clean["annee"] >= 2000) & (decp_clean["annee"] <= 2025)]

# --------------------------
# 1️⃣ Top 10 acheteurs
# --------------------------
top_acheteurs_dash = decp_dash.groupby("acheteur_id").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False).head(10)

print("🔹 Top 10 acheteurs :")
display(top_acheteurs_dash)

# Graphique
plt.figure(figsize=(12,6))
sns.barplot(x=top_acheteurs_dash.index, y=top_acheteurs_dash["total_montant"], color="skyblue", label="Montant total (€)")
sns.lineplot(x=top_acheteurs_dash.index, y=top_acheteurs_dash["nb_marches"], color="red", marker="o", label="Nombre de marchés")
plt.title("Top 10 acheteurs : Montant total vs Nombre de marchés")
plt.xlabel("Acheteur ID")
plt.ylabel("Montant total")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# --------------------------
# 2️⃣ Top 10 secteurs (Code CPV)
# --------------------------
top_cpv = decp_dash.groupby("codeCPV").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False).head(10)

print("\n🔹 Top 10 secteurs (Code CPV) :")
display(top_cpv)

plt.figure(figsize=(12,6))
sns.barplot(x=top_cpv.index, y=top_cpv["total_montant"], color="lightgreen", label="Montant total (€)")
sns.lineplot(x=top_cpv.index, y=top_cpv["nb_marches"], color="orange", marker="o", label="Nombre de marchés")
plt.title("Top 10 secteurs (Code CPV) : Montant total vs Nombre de marchés")
plt.xlabel("Code CPV")
plt.ylabel("Montant total")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# --------------------------
# 3️⃣ Analyse temporelle
# --------------------------
montant_par_annee = decp_dash.groupby("annee")["montant"].sum()
nb_marches_par_annee = decp_dash.groupby("annee")["montant"].count()

print("\n🔹 Montant total par année :")
display(montant_par_annee)

print("\n🔹 Nombre de marchés par année :")
display(nb_marches_par_annee)

plt.figure(figsize=(12,6))
sns.lineplot(x=montant_par_annee.index, y=montant_par_annee.values, marker="o", label="Montant total (€)")
sns.lineplot(x=nb_marches_par_annee.index, y=nb_marches_par_annee.values, marker="o", label="Nombre de marchés")
plt.title("Évolution du montant total et du nombre de marchés par année")
plt.xlabel("Année")
plt.ylabel("Valeur")
plt.legend()
plt.show()

# --------------------------
# 4️⃣ KPIs qualitatifs par année
# --------------------------
kpi_annee = decp_dash.groupby("annee").agg(
    taux_innovant=pd.NamedAgg(column="marcheInnovant", aggfunc=lambda x: (x.str.lower() == "oui").mean()),
    taux_social=pd.NamedAgg(column="considerationsSociales", aggfunc=lambda x: (x.str.lower() != "pas de considération sociale").mean()),
    taux_env=pd.NamedAgg(column="considerationsEnvironnementales", aggfunc=lambda x: (x.str.lower() != "pas de considération environnementale").mean())
)

print("\n🔹 KPIs qualitatifs par année :")
display(kpi_annee)

plt.figure(figsize=(12,6))
sns.lineplot(data=kpi_annee*100, markers=True)
plt.title("Évolution des KPI qualitatifs par année (%)")
plt.ylabel("Pourcentage (%)")
plt.xlabel("Année")
plt.legend(["Innovants", "Critères sociaux", "Critères environnementaux"])
plt.show()
```

```python
# --------------------------
# Mapping CPV → Nom des secteurs
# --------------------------
cpv_mapping = {
    "80530000-8": "Srvc formation professionnelle",
    "30200000-1": "Srvc informatiques",
    "33000000-0": "Fournitures industrielles",
    "79100000-5": "Srvc de conseil",
    "09310000-5": "Produits agricoles",
    "45000000-7": "Trvx de construction générale",
    "64200000-8": "Services postaux",
    "45233140-2": "Trvx préparatoires pour construction",
    "33600000-6": "Equipement mécanique",
    "03121100-6": "Prodts alimentaires spécifiques"
}

# Ajouter une colonne avec le nom du secteur
decp_clean["nom_secteur"] = decp_clean["codeCPV"].map(cpv_mapping)

# Top 10 secteurs par montant total
top_secteurs = decp_clean.groupby("nom_secteur").agg(
    total_montant=pd.NamedAgg(column="montant", aggfunc="sum"),
    nb_marches=pd.NamedAgg(column="montant", aggfunc="count")
).sort_values(by="total_montant", ascending=False).head(10)

# Afficher tableau
print("🔹 Top 10 secteurs avec noms :")
display(top_secteurs)

# Graphique
plt.figure(figsize=(12,6))
sns.barplot(x=top_secteurs.index, y=top_secteurs["total_montant"], color="lightblue", label="Montant total (€)")
sns.lineplot(x=top_secteurs.index, y=top_secteurs["nb_marches"], color="red", marker="o", label="Nombre de marchés")
plt.title("Top 10 secteurs par montant total et nombre de marchés")
plt.xlabel("Nom du secteur")
plt.ylabel("Montant total (€)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
